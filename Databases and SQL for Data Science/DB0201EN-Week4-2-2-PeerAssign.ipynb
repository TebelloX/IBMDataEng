### Problem 6

##### Which schools in Community Areas 10 to 15 are healthy school certified?


```python
school_df = pd.read_csv("Public_Schools.csv")
```

```python
school_df
```

```python
school10to15 = school_df[(school_df["COMMUNITY_AREA_NUMBER"] >= 10) & (school_df["COMMUNITY_AREA_NUMBER"] <= 15)]
school10to15
```

```python
school10to15["HEALTHY_SCHOOL_CERTIFIED"].value_counts()
```

### Problem 7

### Problem 8
